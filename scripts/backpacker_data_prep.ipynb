```python
# Clean up any potential whitespace
cost_df['Country'] = cost_df['Country'].str.strip()
quality_df['Country'] = quality_df['Country'].str.strip()
```

```python
# Clean up whitespace just in case
cost_df['Country'] = cost_df['Country'].str.strip()
quality_latest_df['Country'] = quality_latest_df['Country'].str.strip()

# Run the merge
merged_df = pd.merge(cost_df, quality_latest_df, on='Country', how='inner')

print(f"The final master dataset has {merged_df.shape[0]} rows (countries).")
```
